
## 9.1 Forward selection Models

En la selecci√≥n hacia adelante, comenzamos con un modelo nulo y luego comenzamos a ajustar el modelo con cada caracter√≠stica individual una a la vez y seleccionamos la caracter√≠stica con el valor p m√≠nimo. Ahora ajuste un modelo con dos caracter√≠sticas probando combinaciones de la caracter√≠stica seleccionada anteriormente con todas las dem√°s caracter√≠sticas restantes. Vuelva a seleccionar la funci√≥n con el valor p m√≠nimo. Ahora ajuste un modelo con tres caracter√≠sticas probando combinaciones de dos caracter√≠sticas previamente seleccionadas con otras caracter√≠sticas restantes. Repita este proceso hasta que tengamos un conjunto de caracter√≠sticas seleccionadas con un valor p de caracter√≠sticas individuales menor que el nivel de significancia.

En resumen, los pasos para la t√©cnica de selecci√≥n hacia adelante son los siguientes:

1. Elija un nivel de significancia (por ejemplo, SL = 0.05 con un 95% de confianza).

2. Ajuste todos los modelos de regresi√≥n simple posibles considerando una caracter√≠stica a la vez. Los modelos totales 'n' son posibles. Seleccione la caracter√≠stica con el valor p m√°s bajo.

3. Ajuste todos los modelos posibles con una caracter√≠stica adicional agregada a las caracter√≠sticas seleccionadas anteriormente.

4. Nuevamente, seleccione la funci√≥n con un valor p m√≠nimo. si $p_v <alpha$, vaya al Paso 3; de lo contrario, finalice el proceso.


Seleccionamos este tipo de m√©todo pues es menos costoso computacionalmente hablando , dado que se inicia con un modelo nulo y van ingresando variables a medida que tanto su significancia como su aporte hacia la variabilidad del modelo lo indiquen ( usualmente un test F)

```python
df.dtypes
```

```python
# Creamos un nuevo df con todos los features y dejamos como target la variable binaria asociada a 0 como aprobado y 1 como reprobado

X = df.drop(['AGNO','DGV_RBD','NOM_RBD','COD_PRO_RBD','MRUN','FEC_NAC_ALU','LET_CUR','NOM_REG_RBD_A','NOM_COM_RBD','NOM_COM_ALU','SIT_FIN_R','NOM_DEPROV_RBD','COD_RAMA','COD_SEC','COD_ESPE','aprobacion_R'], axis=1)    # feature matrix
y = df['aprobacion_R']               # target feature
```

```python
X.dtypes
```

```python
X = X.dropna()
y = y.dropna()
```

```python
# Esta funci√≥n nos permitir√° realizar una seleccci√≥n de variables usando el m√©todo Forward

import statsmodels.api as sm
def forward_selection(data, target, significance_level=0.01):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<significance_level):
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features
```

```python
#selecciono por el m√©todo forward las variables m√°s significativas
forward_selection(data=X,target=y,significance_level=0.01)
```

```python
 from sklearn.ensemble import RandomForestClassifier
 from sklearn.model_selection import train_test_split
 from sklearn import metrics
```

```python
# Modelo Random Forest 1
#========================================

# Separamos en entrenamiento/test en razon 80/20 %
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=1)
# Creamos un modelo Random Forest con parametros por defecto
modelo = RandomForestClassifier(random_state=1)
modelo.fit(X_train, y_train)
# Obtenemos las predicciones del modelo con X_test
preds = modelo.predict(X_test) 
```

```python
#gr√°ficamos la matriz de confusi√≥n
plt.figure(figsize=(10,5))
metrics.plot_confusion_matrix(modelo, X_test, y_test, display_labels=['Negative', 'Positive'])
plt.show()
```

```python
#revisamos las metricas
print(metrics.classification_report(y_test, preds))
```

## 9.3.2 Cross-Validation

Cross-Validation es una de las formas m√°s eficientes de interpretar el rendimiento del modelo. Garantiza que el modelo se ajuste con precisi√≥n a los datos y tambi√©n verifica cualquier sobreajuste. Es el proceso de evaluar c√≥mo los resultados de un an√°lisis estad√≠stico se generalizar√°n a un conjunto de datos independiente.

El sobreajuste ocurre cuando el modelo encaja perfectamente en el conjunto de entrenamiento y tiene un desempe√±o deficiente en el conjunto de prueba. Esto significa que el modelo aprender√° bien los datos de entrenamiento pero no generalizar√° sobre muestras no vistas.

Existen diferentes t√©cnicas de Cross-Validation, tales como:

*   Cross-Validation de K-Fold
*   Cross-Validation estratificada de K-Fold
*Cross-Validation basada en retenci√≥n
*Cross-Validation cruzada Leave-One-Out
*Cross-Validation cruzada del grupo K-Fold



La idea general detr√°s de la validaci√≥n cruzada es que dividimos los datos de entrenamiento en algunas partes. Elegimos algunas de estas partes para entrenar y el resto para probar el modelo. Las diferentes t√©cnicas de validaci√≥n cruzada se basan en c√≥mo particionamos los datos.

```python
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
# prepare the cross-validation procedure
cv = KFold(n_splits=10, random_state=1, shuffle=True)
# create model
model_rf = RandomForestClassifier(random_state=1)
# evaluate model
scores = cross_val_score(model_rf, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

# 9.3.2 Feature Selection


Al construir un modelo de aprendizaje autom√°tico en la vida real, es casi raro que todas las variables en el conjunto de datos sean √∫tiles para construir un modelo. Agregar variables redundantes reduce la capacidad de generalizaci√≥n del modelo y tambi√©n puede reducir la precisi√≥n general de un clasificador. Adem√°s, agregar m√°s y m√°s variables a un modelo aumenta la complejidad general del modelo.

De acuerdo con la Ley de Parsimonia, la mejor explicaci√≥n de un problema es aquella que involucra la menor cantidad de suposiciones posibles. Por lo tanto, la selecci√≥n de caracter√≠sticas se convierte en una parte indispensable de la construcci√≥n de modelos de machine learning

En el Aprendizaje No Supervizado: estas t√©cnicas se pueden utilizar para datos no etiquetados.

Desde un punto de vista taxon√≥mico, estas t√©cnicas se clasifican en:

A. M√©todos de filtrado: Los m√©todos de filtro recogen las propiedades intr√≠nsecas de las caracter√≠sticas medidas a trav√©s de estad√≠sticas univariadas en lugar del rendimiento de la validaci√≥n cruzada. Estos m√©todos son m√°s r√°pidos y menos costosos desde el punto de vista computacional que los m√©todos de envoltura. Cuando se trata de datos de alta dimensi√≥n, es computacionalmente m√°s barato usar m√©todos de filtro.

B. M√©todos de envoltura (Wrapper): Los contenedores requieren alg√∫n m√©todo para buscar en el espacio de todos los posibles subconjuntos de caracter√≠sticas, evaluando su calidad aprendiendo y evaluando un clasificador con ese subconjunto de caracter√≠sticas. El proceso de selecci√≥n de funciones se basa en un algoritmo de aprendizaje autom√°tico espec√≠fico que intentamos adaptar a un conjunto de datos determinado. Sigue un enfoque de b√∫squeda codicioso al evaluar todas las posibles combinaciones de caracter√≠sticas contra el criterio de evaluaci√≥n. Los m√©todos de envoltorio generalmente dan como resultado una mejor precisi√≥n predictiva que los m√©todos de filtro.

C. M√©todos incorporados : Estos m√©todos abarcan los beneficios de los m√©todos de envoltura y filtro, al incluir interacciones de caracter√≠sticas pero tambi√©n manteniendo un costo computacional razonable. Los m√©todos embebidos son iterativos en el sentido de que se encargan de cada iteraci√≥n del proceso de entrenamiento del modelo y extraen cuidadosamente aquellas caracter√≠sticas que contribuyen m√°s al entrenamiento para una iteraci√≥n particular.





```python
# Clase para realizar componentes principales
from sklearn.decomposition import PCA

# Estandarizador (transforma las variables en z-scores)
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler() # Creamos el estandarizador para usarlo posteriormente
```

```python
# Features que vamos a considerar
features = ['GEN_ALU',
 'COD_ENSE',
 'ASISTENCIA',
 'COD_ENSE2',
 'PROM_GRAL',
 'COD_GRADO',
 'EDAD_ALU',
 'COD_DES_CUR',
 'ESTADO_ESTAB',
 'RBD',
 'COD_REG_RBD',
 'COD_DEPE',
 'COD_DEPE2',
 'COD_JOR',
 'COD_COM_RBD',
 'COD_TIP_CUR',
 'COD_COM_ALU',
 'COD_MEN']

# Matriz de numpy de los features
X = df[features].to_numpy()

print('Dimensiones de la matriz de features: {}'.format(X.shape))
```

Antes de hallar las componentes principales, buscamos estandarizar las variables (rest√°ndole la media y dividiendo por la desviaci√≥n est√°ndar), lo que permite eliminar una variabilidad ficticia de los datos debido al simple hecho de estar medidas en diferentes unidades:

```python
# Ajustamos el estandarizador
std_scale.fit(X)

# Aplicamos el estandarizador y obtenemos la matriz de features escaleados
X_scaled = std_scale.transform(X)
```

```python
# Creaci√≥n del modelo. Si el n√∫mero de componentes no se espec√≠fica, 
# se obtienen tantas componentes principales como features en nuestro dataset.
pca = PCA(n_components=None)
```

```python
# Ajustamos el modelo a los datos escaleados
pca.fit(X_scaled)

# Obtenemos la descripci√≥n de los datos en el espacio de componentes principales
X_pca = pca.transform(X_scaled)

print('Dimensiones de la matriz en componentes principales: {}'.format(X_pca.shape))
print(X_pca)
```

Fracci√≥n de varianza explicada
Uno de los detalles m√°s interesantes de PCA es que las componentes principales est√°n ordenadas de mayor a menor en el sentido de informaci√≥n que aportan. As√≠, la primer componente es la m√°s informativa, la segunda aporta menos que la primera pero m√°s que las otras, y as√≠ siguiendo. Hablar de informaci√≥n en este contexto es sin√≥nimo de dar la fracci√≥n de varianza explicada por cada componente. Podemos ver cu√°nto aporta cada componente, como tambi√©n cu√°nto es el acumulado de todas las componentes:

```python
# con .explained_variance_ratio_ vemos la fracci√≥n de informaci√≥n que aporta cada componente
evr = pca.explained_variance_ratio_

# Graficamos la fracci√≥n de varianza que aporta cada componente
# y la informaci√≥n acumulada
fig, ax = plt.subplots(1, 2, figsize = (12, 4))

ax[0].plot(range(1, len(evr) + 1), evr, '.-', markersize = 20)
ax[0].set_ylabel('Fracci√≥n de varianza explicada')
ax[0].set_xlabel('N√∫mero de componente principal')

# Calculamos el acumulado con la funci√≥n cumsum de numpy 
varianza_acumulada = np.cumsum(evr)

ax[1].plot(range(1, len(evr) + 1), varianza_acumulada, '.-', markersize = 20)
ax[1].set_ylabel('Fracci√≥n acumulada de varianza explicada')
ax[1].set_xlabel('Cantidad de componentes principales')
```

El gr√°fico de la izquierda nos dice qu√© fracci√≥n de informaci√≥n aporta cada componente. Lo que es importante notar es que cada componente aporta menos informaci√≥n que la anterior, por lo que las mismas est√°n ordenadas de m√°s a menos informativas.

El gr√°fico de la derecha es m√°s interesante: lo que nos dice es que tomando solamente las 5 primeras componentes cubrimos m√°s del 60% de la informaci√≥n, con 10, un poco m√°s del 90%, y as√≠. Cuando tomamos tantas componentes como features en nuestro dataset original, no hay p√©rdida de informaci√≥n.

```python
print('Features = {}'.format(features))
print('PCA1 = {}'.format(pca.components_[0]))
print('PCA2 = {}'.format(pca.components_[1]))
print('PCA3 = {}'.format(pca.components_[2]))
print('PCA4 = {}'.format(pca.components_[3]))
print('PCA5 = {}'.format(pca.components_[4]))
```

Visualizaci√≥n de los datos en el espacio reducido
Otra ventaja que tiene PCA es que podemos utilizarlo para visualizar los datos originalmente multidimensionales en un espacio de dimensi√≥n reducida, idealmente en dos dimensiones para mayor facilidad en la interpretaci√≥n. Sin embargo, tener presente que el espacio reducido tiene solo una parte de la informaci√≥n y que si esta es poca, la visualizaci√≥n puede no ser adecuada.

Vemos los datos en el espacio de las dos primeras componentes principales:

```python
fig, ax = plt.subplots(figsize = (10, 6))

# Hacemos un scatter de los datos en las dos primeras componentes
ax.scatter(X_pca[:,0], X_pca[:,1], alpha = 0.65)

ax.set_xlabel('Primer componente principal')
ax.set_ylabel('Segunda componente principal')
```

```python
# Matriz de features en el espacio de componentes principales
# Simplemente la matriz traspuesta de las componentes principales en el espacio de features
features_pc = pca.components_.T

print('Descripci√≥n de la variable Asistencia en las componentes principales:')
print(features_pc[2])
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

# Hacemos un scatter de los datos en las dos primeras componentes
ax.scatter(X_pca[:,0], X_pca[:,1], alpha = 0.65)

# Hacemos el grafico de las flechas indicando las direcciones de los features originales

sf = 4 # Factor de escala solo a efecto de estirar 
       # las flechas para que sean visibles (solo nos importa la direcci√≥n en realidad)

# Recorremos cada feature
for i in range(len(features)):

  # Creamos una flecha que vaya del origen y apunte en la direcci√≥n de los features
  ax.arrow(0, 0, sf * features_pc[i][0], sf * features_pc[i][1], width = 0.1, color = 'r', alpha = 0.65)

  # Indicamos con texto a qu√© feature corresponde cada flecha
  ax.text(sf * features_pc[i][0], sf * features_pc[i][1], s = features[i], fontdict= {'color': 'k', 'size': 15})

ax.set_xlabel('Primer componente principal')
ax.set_ylabel('Segunda componente principal')
```

La descripci√≥n de nuestros features en el espacio reducido de las primeras componentes principales puede ser utilizado como input para diferentes modelos. En particular, para el caso del modelo de regresi√≥n lineal, recordemos que las variables independientes no deb√≠an ser colineales (es decir, una no deb√≠a poder escribirse como una combinaci√≥n lineal de otras). Dado que las componentes principales son por construcci√≥n ortogonales, son por defecto buenas candidatas para formar parte de un modelo de regresi√≥n.
