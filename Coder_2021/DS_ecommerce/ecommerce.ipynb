Segmentación de clientes: 
Utilizando el algoritmo k-means agruparemos a los clientes (identificándolos mediante el client_id) en clusters de acuerdo a dos variables
- recurrencia de compra (cantidad de transacciones que realizaron)
- cantidad de dinero que gastaron (suma del precio de los productos que adquirieron)

```python
from sklearn.cluster import KMeans #importamos la librería

#Armamos un dataset filtrado con los datos que necesitamos, eliminando los registros con precio nulo
df_k = df.dropna(subset=['client_price']).groupby('client_id')['client_price'].agg(spent ='sum', transactions= 'count')
df_k=df_k.reset_index()
df_k=df_k[['spent','transactions']]
df_k.sort_values(by='transactions', ascending=False).head()
```

```python
#Estandarización de datos (para que manejen la misma escala)
scaler = StandardScaler()
df_k_std = scaler.fit_transform(df_k)
```

```python
#Aplicamos el elbow method para encontrar el número óptimo de clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df_k_std)
    wcss.append(kmeans.inertia_)

# Ploteamos el resultado para conocer cuantos clusters serán mejor analizar
plt.figure(figsize = (8,6))
plt.plot(range(1, 11), wcss, marker = 'o', linestyle = '--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means Clustering',fontsize=18)
plt.show()
```

De acuerdo al gráfico obtenido con el Elbow Method determinamos que lo óptimo es armar 4 clústers

```python
#Creamos el k-Means con 4 grupos / clústers
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 47)
y_kmeans = kmeans.fit_predict(df_k_std)
y_kmeans
```

```python
#Agregamos la columna del clúster (pred) que determinó el algoritmo para cada registro
df_k['pred']=y_kmeans
df_k.sort_values(by='transactions', ascending=False)
```

```python
#Analizamos los promedios para cada grupo
df_k.groupby(['pred']).mean().round()
```

```python
#Vemos la cantidad de clientes que quedó en cada grupo
df_k['pred'].value_counts()
```

```python
# Graficamos los resultados del Algoritmo K-means
x_axis = df_k['spent']
y_axis = df_k['transactions']
plt.figure(figsize = (10, 8))
sns.scatterplot(x_axis, y_axis, hue = df_k['pred'], palette = ['g', 'r', 'c','m'])
plt.title('Segmentation K-means',fontsize=16)
plt.show()
```

En conclusión:
- El grupo 0 contiene la mayor cantidad de clientes, que en promedio solo compraron  1 vez y gastaron poco
- El grupo 1 contiene clientes que compran esporádicamente o compraron una vez pero gastaron más que la mayoría
- El grupo 2 contiene clientes recurrentes que gastan montos considerables de dinero.
- El grupo 3 contiene solo 2 clientes que compraron 71 veces en promedio gastando montos considerables de dinero. Se debería analizar si corresponden a clientes mayoristas.

Calcularemos el índice de Silhouette para evaluar si la cantidad de grupos elegidos (4) es el más óptimo

```python
#Cálculo del índice de Silouette
from sklearn.metrics import silhouette_score
score = silhouette_score(df_k_std, kmeans.labels_, metric='euclidean')
print('Silhouetter Score: %.3f' % score)
```

```python
#Visualizaciones gráficas del índice de Silhouette para K= 2, 3, 4 y 5
from yellowbrick.cluster import SilhouetteVisualizer

fig, ax = plt.subplots(2, 2, figsize=(15,8)) #Definimos el tamaño del gráfico
for i in [2, 3, 4, 5]:
    
    #Creamos diferentes instancias de kmeans para las distintas cantidades de clusters
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=50, random_state=42)
    q, mod = divmod(i, 2)

    #Visualizamos el índice Silhouette con yellowbrick
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(df_k_std)
```

Conclusión: Podemos confirmar que k = 4 es el valor óptimo para la clusterización, tal como nos indicó el Elbow method, ya que se logra una mejor distribución de los datos entre grupos. 5 grupos también podría ser un valor a considerar, ya que se distribuyen aún mejor los datos, pero como contrapartida se generan 2 grupos casi sin observaciones

Hyperparameter Optimization: Partiendo del modelo que mejor se adapta a los datos (Random Forest) intentaremos ajustar sus parámetros para conocer si es posible mejorar su precisión.

```python
# Numero de Árboles del Random Forest
n_estimators = [int(x) for x in np.linspace(start = 5, stop = 20, num =5)]
# Numero de caracterísiticas a considerar en cada split
max_features = ['auto', 'sqrt']
# Número máximo de niveles por Arbol.
max_depth = [2,5]
# Número mínimo de muestras necesarias para dividir un nodo.
min_samples_split = [2,5]
# Número mínimo de muestras necesarias en cada nodo hoja.
min_samples_leaf = [2,5]
# Método de selección de muestras para entrenar cada árbol.
bootstrap = [True, False]
```

```python
from sklearn.model_selection import GridSearchCV

# Creamos grilla de parámetros
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}

#Realizamos el Grid Search CV
RF_grid = GridSearchCV(RF, param_grid, cv = 5, n_jobs = -1, scoring = 'precision')

#Entrenamos
RF_grid.fit(X_train,y_train)

#Devolvemos los mejores parámetros a utilizar
print("Best parameters: ")
print(RF_grid.best_params_)
```
