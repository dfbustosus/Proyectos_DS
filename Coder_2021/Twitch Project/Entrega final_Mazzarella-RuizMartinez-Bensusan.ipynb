```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

df = pd.read_csv('TW2.csv')
df.head()
```

```python
datos = pd.DataFrame(df)
```

```python
#Analizamos la variable target y su frecuencia
df['genero_bin'].value_counts()
```

```python
#Definimos X e y
X = df[['ID_jugador','ranking_jugador','tiempo_vistas','tiempo_transmision','max_vistas','Height','Weight','Index']]
y = df['genero_bin']
```

```python
X
```

```python
y
```

```python
#Separamos en Train y Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
!pip install lightgbm 
```

```python
import lightgbm as lgb #pip install lightgbm
clf = lgb.LGBMClassifier()
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
```

```python
#Accuracy
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_pred, y_test)
print('LightGBM Model accuracy score: {0:0.4f}'.format(accuracy_score(y_test, y_pred)))
```
